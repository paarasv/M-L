```python
import numpy as np 
import matplotlib.pyplot as plt
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
def gd(x,y):
    m_curr=0
    b_curr=0
    n=len(x)
    iterations=1000
    rate=0.01
    plt.scatter(x,y,color="r",alpha=0.5)
    
    for i in range (iterations):
        yp = m_curr*x + b_curr
        cost=((1/n)*sum([val**2 for val in(y-yp)]))
        plt.plot(x,yp,color="green")
        md = -(2/n)*sum(x*(y-yp))
        yd = -(2/n)*sum(y-yp)
        m_curr = m_curr - rate * md
        b_curr = b_curr - rate * yd 
        print("m_curr",m_curr,"b_curr",b_curr)
        
gd(x,y)
              
    
```
